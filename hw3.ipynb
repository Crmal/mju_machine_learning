```python
import numpy as np
import matplotlib.pyplot as plt
datas = np.loadtxt('weightHeight.csv', delimiter=",")

#남녀 데이터 분할
mans_data = datas[datas[:,0] == 0] # datas에서 첫번째 행이 0인것을 추출(남자의 데이터)
womans_data = datas[datas[:,0] == 1] # datas에서 첫번째 행이 0인것을 추출(여자의 데이터)

#남녀 키, 몸무게 데이터 분할
m1
womans_data_x = womans_data[:,2] #여자의 키
womans_data_y = womans_data[:,1] #여자의 몸무게

#선형회귀분석에 필요한 데이터 추출 
mans_k, mans_d = np.polyfit(mans_data_x, mans_data_y,1) #남자
mans_pred = mans_k * mans_data_x + mans_d 

womans_k, womans_d = np.polyfit(womans_data_x, womans_data_y,1) #여자
womans_pred = womans_k * womans_data_x + womans_d 

#데이터 표현
plt.plot(mans_data_x,mans_data_y,'.')#남자
plt.plot(mans_data_x, mans_pred)
```
